## 本节要求 /Requirements/

* 掌握线性齐次微分方程组的解的性质及代数结构.
* 掌握线性非齐次微分方程组的解的代数结构,理解常数变易法的基本思想.

对于线性微分方程组

$$
x'=A(t)x+f(t)
$$

如果 $ f(t)\neq 0$则 
$$
x'=A(t)x+f(t)
$$
称为**非齐次线性**的.


如果 $f(t) = 0$ 则 
$$
x'=A(t)x
$$
称为**齐次线性**的.

若 $A(t)$ 为常数矩阵,则称为**常系数线性方程组**.

$$
x'=Ax
$$

### 5.2.1 齐线性微分方程组

$$
x'=A(t)x
$$

**定理2**(叠加原理) 如果 $u(t)$ 和 $v(t)$ 是

$$
x'=A(t)x
$$

的解, 则它们的线性组合 $\alpha u(t)+\beta v(t)$ 也是 $x'=A(t)x$ 的解.

**证明:** 

$$
\begin{aligned}
{[\alpha u(t)+\beta v(t)]'}
&=\alpha u(t)'+\beta v(t)'\\
&=\alpha A(t)u(t)+\beta A(t) v(t)\\
&=A(t)[\alpha u(t)+\beta v(t)]\\
\end{aligned}
$$

如果 $x_1(t),x_2(t),\dotsi,x_n(t)$ 是 $x'=A(t)x$ 的解,则

$$
c_1x_1(t)+c_2x_2(t)+\dotsi+c_nx_n(t)
$$

也是 $x'=A(t)x$ 的解.

可验证 $\quad$ 
$$
x_1(t)=
\left[
\begin{matrix}
\sin t\\
\cos t\\
\end{matrix}
\right],
x_2(t)=
\left[
\begin{matrix}
\cos t\\
-\sin t\\
\end{matrix}
\right]
$$

是方程组

$$
x'=
\left[
\begin{matrix}
0&1\\
-1&0\\
\end{matrix}
\right]x
$$

的解,则

$$
x(t)=
c_1
\left[
\begin{matrix}
\sin t\\
\cos t\\
\end{matrix}
\right]+c_2
\left[
\begin{matrix}
\cos t\\
-\sin t\\
\end{matrix}
\right]
$$

也是方程组的解.

### 基本概念 /Basic Concept/

定义在区间 $a \leq t \leq b$ 上的向量函数

$$
x_1(t),x_2(t),\dotsi,x_m(t)
$$

是 **线性相关**的
线性相关的 , 如果存在不全为零的常数

$c_1,c_2,\dotsi,c_m,$ 使得等式

$$
c_1x_1(t)+c_2x_2(t)+\dotsi+c_mx_m(t)\equiv0,a\leq t \leq b
$$

成立;否则,

$$
x_1(t),x_2(t),\dotsi,x_m(t)
$$

为**线性无关**的.

**例**

$$
\left[
\begin{matrix}
1\\
0\\
\vdots\\
0\\
\end{matrix}
\right],
\left[
\begin{matrix}
0\\
t\\
\vdots\\
0\\
\end{matrix}
\right],\left[
\begin{matrix}
0\\
0\\
\vdots\\
t^k\\
\end{matrix}
\right],
$$

**线性无关**
$(-\infty<t<\infty)$

设有 $n$ 个定义在区间 $a\leq t \leq b$ 上的向量函数

$$
x_1(t)=
\left[
\begin{matrix}
x_{11}(t)\\
x_{21}(t)\\
\vdots\\
x_{n1}(t)\\
\end{matrix}
\right],
x_2(t)=
\left[
\begin{matrix}
x_{12}(t)\\
x_{22}(t)\\
\vdots\\
x_{n2}(t)\\
\end{matrix}
\right],
\dotsi
x_n(t)=
\left[
\begin{matrix}
x_{1n}(t)\\
x_{2n}(t)\\
\vdots\\
x_{nn}(t)\\
\end{matrix}
\right],
$$

由这 $n$ 个向量函数构成的行列式,

$$
W[x_1(t),x_2(t),\dotsi,x_n(t)]\equiv W(t)\equiv
det\left[
\begin{matrix}
x_{11}(t)&x_{12}(t)&\dots&x_{1n}(t)\\
x_{21}(t)&x_{22}(t)&\dots&x_{2n}(t)\\
\vdots&\vdots&\vdots&\vdots&\\
x_{n1}(t)&x_{n2}(t)&\dots&x_{nn}(t)\\
\end{matrix}
\right],
$$

称为这些向量函数的**伏朗斯基行列式**


**定理$3$** 如果向量函数 $x_1(t),x_2(t),\dotsi,x_n(t)$ 在区间 $a\leq t \leq b$ 上线性相关, 则它们的伏朗斯基行列式

$$
W(t)\equiv0,a \leq t \leq b
$$

**证明** 由假设,存在不全为零的常数 $c_1,c_2,\dotsi,c_n$ 使得

$$
c_1x_1(t)+c_2x_2(t)+\dotsi+c_nx_n(t)\equiv0,a\leq t \leq b
$$

$$
\begin{cases}
c_1x_{11}(t)+c_2x_{12}(t)+\dotsi+c_nx_{1n}(t)=0\\
c_1x_{21}(t)+c_2x_{22}(t)+\dotsi+c_nx_{2n}(t)=0\\
\dots\dots\dots\dots\\
c_1x_{n1}(t)+c_2x_{n2}(t)+\dotsi+c_nx_{nn}(t)=0\\
\end{cases}
$$

其系数行列式恰是 $W(t)$

$$
W(t)\equiv0,a\leq t \leq b
$$

**定理$4$**如果
$$
x'=A(t)x
$$

的解 $x_1(t),x_2(t),\dotsi,x_n(t)$ 线性无关,那么 , 它们的伏朗斯基行列式

$$
W(t)\neq0,a\leq t \leq b
$$

**证明 **用反证法

设有某一个 $t_0$ ,$a\leq t_0 \leq b$ 使得 $W(t_0)=0$

考虑下面的齐次线性代数方程组

$$
c_1x_1(t_0)+c_2x_2(t_0)+\dotsi+c_nx_n(t_0)=0
$$

它的系数行列式 $W(t_0)=0$ ,所以

$$
c_1x_1(t_0)+c_2x_2(t_0)+\dotsi+c_nx_n(t_0)=0
$$

有非零解,$\tilde{c_1},\tilde{c_n},\dotsi,\tilde{c_n}$

$$
\tilde{c_1}x_1(t_0)+\tilde{c_2}x_2(t_0)+\dotsi+\tilde{c_n}x_n(t_0)=0
$$

以这个非零解作向量函数

$$
x(t)=\tilde{c_1}x_1(t)+\tilde{c_2}x_2(t)+\dotsi+\tilde{c_n}x_n(t)
$$

易知 $x(t)$ 是

$$
x'=A(t)x
$$

的解 , 且满足初始条件

$$
x(t_0)=0
$$

而在 $a \leq t \leq b $ 上恒等于零的向量函数 $0$ 也是

$$
x'=A(t)x
$$

满足初始条件
$$
x(t_0)=0
$$

的解.

由解的唯一性，知道 $x(t)\equiv0$ 即

$$
\tilde{c_1}x_1(t)+\tilde{c_2}x_2(t)+\dotsi+\tilde{c_n}x_n(t)=0,a \leq t \leq b
$$

因为 $\tilde{c_1},\tilde{c_2},\dotsi,\tilde{c_n}$ 不全为零, 这就与$x_1(t),x_2(t),\dots,x_n(t)$ 线性无关矛盾.

**结论** 由
$$
x'=A(t)x
$$
的解$x_1(t),x_2(t),\dots,x_n(t)$作成的伏朗斯基行列式$W(t)$或者恒等于零, 或者恒不等于零.

**定理$5$**
$$

**定理$6$**如果$x_1(t),x_2(t),\dotsi,x_n(t)$是

$$

$$

上式看作是以$c_1,c_2,\dots,c_n$为未知量的线性代数方程组,系数行列式就是

$Ｗ(t_0)$,因为$x_1(t),x_2(t),\dotsi,x_n(t)$线性无关,则$W(t_0)\neq 0$,
