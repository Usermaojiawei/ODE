# 5.2.2 非齐次线性微分方程组

本节讨论的是非齐次线性微分方程组

$$
x'=A(t)x+f(t)
$$

的解的结构问题,$A(t)$ 是区间 $a\leq t \leq b$ 上的已知 $n$ 阶连续矩形,$f(t)$ 是区间 $a\leq t \leq b$ 的 $n$ 维连续列向量.

### 性质1 

如果 $\varphi (t)$ 是此方程的解,$\psi$ 是此方程对应的齐次线性微分方程组 $x'=A(t)x$ 的解. 则 $\varphi + \psi$ 是此方程的解.

### 性质2

如果 $\tilde{\varphi}(t)$ 和 $\bar{\varphi}(t)$ 是上方程的两个解, 则 $\tilde{\varphi}(t)-\bar{\varphi}(t)$ 是齐次线性微分方程组 $x'=A(t)x$ 的解. 

### 定理7

设 $\phi(t)$ 是方程 $x'=A(t)x$ 的基解矩阵,$\bar{\varphi}(t)$ 是

$$
x'=A(t)x+f(t)
$$

的某一解,则 $x'=A(t)x$ 的任一解 $\varphi(t)$ 都可以表示为

$$
\varphi(t)=\psi(t)c-\bar{\varphi}(t)
$$
这里 $c$ 是确定的常数列向量.

**证明 ： ** 由性质2我们知道 $\tilde{\varphi}(t)-\bar{\varphi}(t)$ 是齐次线性微分方程组 $x'=A(t)x$ 的解.因此存在常数列向量 $c$,使得

$$
\varphi(t)-\bar{\varphi}(t)=\psi(t)c\\
\varphi(t)=\bar{\varphi}(t)+\psi(t)c
$$

为了寻求

$$
x'=A(t)x+f(t)
$$

的通解, 只要知道此方程对应的齐次线性微分方程组的**基解矩阵**和自身的一个解即可.已知齐次线性微分方程组的基解矩阵 $\psi(t)$, 即可用**常数变易法**求非齐次线性微分方程组的**特解**$\varphi(t)$.



假设非齐次线性微分方程组

$$
x'=A(t)x+f(t)
$$

存在形如 $\varphi(t)=\phi(t)c(t)$ 的解, 则 
$$
\phi'(t)c(t)+\phi(t)c'(t)=A(t)\phi(t)c(t)+f(t)
$$

而

$$
\phi'(t)=A(t) \phi(t) \\
\phi (t) c'(t)=f(t) \\
c'(t)=\phi ^{-1}(t) f(t)\\
c(t)=\int_{t_0}^t\phi^{-1}(s)f(s)\mathrm ds,\quad t_0,t\to[a,b]\\
c(t_0)=0
$$
这样
$$
\varphi(t)=\phi(t)\int_{t_0}^t\phi^{-1}(s)f(s)\mathrm ds
$$

如果非齐次方程有一个 $\varphi(t)=\phi(t)c(t)$ 的解, 则 $\phi (t)$ 由$\varphi(t)=\phi(t)\int_{t_0}^t\phi^{-1}(s)f(s)\mathrm ds$ 决定. 反之, 易证 $\varphi(t)$ 一定是非齐次方程的解.

$$
\begin{align}
\varphi(t) &= \phi(t)\int_{t_0}^t\phi^{-1}(s)f(s)\mathrm ds\\
\varphi'(t) &= \phi'(t)\int_{t_0}^t\phi^{-1}(s)f(s)\mathrm ds+\phi(t)\phi^{-1}(t)f(t)\\
\varphi'(t) &= A(t)(\phi(t)\int_{t_0}^t \phi ^{-1}(s)f(s)\mathrm ds)+f(t)\\
\varphi'(t) &= A(t)\varphi(t)+f(t)
\end{align}
$$

### 定理8

如果 $\phi(t)$ 是齐次线性微分方程组 $x'=A(t)x$ 的基解矩阵, 则向量函数

$$
\varphi(t) = \phi(t)\int_{t_0}^t \phi^{-1}(s)f(s) \mathrm ds
$$

是非齐次线性微分方程组

$$
x'=A(t)x+f(t)
$$

的解, 且满足初始条件 $\varphi(t_0)=0$.满足初始条件 $\varphi(t_0)=\eta$ 的解是

$$
\varphi (t) =\psi(t)\psi^{-1}(t)\eta+\psi(t)\int_{t_0}^t \psi^{-1}(s)f(s) \mathrm ds
$$

此方程的通解为

$$
\varphi (t) =\psi (t) c+\psi (t)\int_{t_0}^t \psi ^{-1}(s)f(s) \mathrm ds
$$

### 常数变易法/Analytic of Unknow Function Method/

$$
\varphi(t) = \Phi(t)c(t) =x_1(t)c_1(t)+x_2(t)c_2(t)+\dots+x_n(t)c_n(t)\\
\Phi(t)c'(t)= f(t)
$$

$$
\begin{bmatrix}
x_11(t) & x_12(t) & \cdots x_1n(t) \\
x_21(t) & x_22(t) & \cdots x_2n(t) \\
\cdots & \cdots & \cdots\\
x_n1(t) & x_n2(t) & \cdots x_nn(t) 
\end{bmatrix}
\begin{bmatrix}
c'_1(t) \\
c'_2(t) \\
\cdots\\
c'_n(t)
\end{bmatrix}
=
\begin{bmatrix}
f_1(t) \\
f_2(t) \\
\cdots\\
f_n(t) 
\end{bmatrix}
$$
$$
\tilde{W}=det
\begin{bmatrix}
x_11(t) & \cdots & f_1(t) & x_1n(t) \\
x_21(t) & \cdots & f_2(t) & x_2n(t) \\
\cdots  & \cdots & \cdots & cdots \\
x_n1(t) & \cdots & f_n(t) & x_nn(t) 
\end{bmatrix}
$$

$$
\begin{align}
c'_k(t) & =\frac{\tilde{W_k(t)}}{W(t)} \quad k=1,2,\cdots,n\\
c_k(t) & =\int_{t_0}^t\frac{\tilde{W_k(t)}}{W(t)}\mathrm ds \quad k=1,2,\cdots,n
\end{align}
$$

$$
\varphi(t) = \Phi(t)c(t) =x_1(t)c_1(t)+x_2(t)c_2(t)+\dots+x_n(t)c_n(t)\\
\varphi(t)=\sum_{k=1}^n x_k(t)\int_{t_0}^t\frac{\tilde{W_k(t)}}{W(t)}\mathrm ds
$$

是非齐次线性微分方程组

$$
x'=A(t)x+f(t)
$$

的解.
应用到 $n$ 阶线性方程

$$
x^{(n)}+a_1(t)x^{(n-1)}+\cdots+a_n(t)x=0\\
x^{(n)}+a_1(t)x^{(n-1)}+\cdots+a_n(t)x=f(t)
$$

### 推论3              

如果 $a_1(t),a_2(t),\cdots,a_n(t),f(t)$ 是区间 $a\leq t\leq b$ 上的连续函数,$x_1(t),x_2(t),\cdots,x_n(t)$ 是对应齐次方程的基本解组, 那么, 非齐次线性方程 

$$
x^{(n)}+a_1 (t) x^{(n-1)}+\cdots+a_n (t) x=f(t)
$$ 

满足初始条件

$$
\varphi(t_0)=0,\varphi'(t_0)=0,\cdots,\varphi^{(n-1)}(t_0)=0 \quad t_0\to [a,b]
$$

的解为

$$
\varphi(t)=\sum_{k=1}^n x_k(t)\int_{t_0}^t{\frac{W_k[x_1(s),x_2(s),\cdots,x_n(s)]}{W[x_1(s),x_2(s),\cdots,x_n(s)]}}f(s)\mathrm ds
$$

$$
W(t)=det 
\begin{bmatrix}
x_1(t)  & x_2(t)  & \cdots & x_n(t) \\
x'_1(t) & x'_2(t) & \cdots & x'_n(t) \\
\cdots  & \cdots  & \cdots \\
x^{(n-1)}_1(t) & x^{(n-1)}_2(t) & \cdots & x^{(n-1)}_n(t)
\end{bmatrix}
$$

$$
\begin{bmatrix}
x_11(t) & \cdots & 0 & x_1n(t) \\
x_21(t) & \cdots & 0 & x_2n(t) \\
\cdots  & \cdots & \cdots & cdots \\
x_n1(t) & \cdots & 0 & x_nn(t) 
\end{bmatrix}
$$

$x^{(n)}+a_1(t)x^{(n-1)}+\cdots+a_n(t)x=f(t)$ 的常数变易公式是

$$
\varphi(t)=\sum_{k=1}^n x_k(t)\int_{t_0}^t{\frac{W_k[x_1(s),x_2(s),\cdots,x_n(s)]}{W[x_1(s),x_2(s),\cdots,x_n(s)]}}f(s)\mathrm ds
$$

方程的通解可以表示为

$$
x=c_1x_1(t)+c_2x_2(t)+\cdots+c_nx_n(t)+\varphi(t)
$$


当 $n=2$ 时, 

$$
\varphi(t)=x_1(t)\int_{t_0}^t\frac{W_1[x_1(s),x_2(s)]}{W[x_1(s),x_2(s)]}f(s)\mathrm ds
+x_2(t)\int_{t_0}^t\frac{W_2[x_1(s),x_2(s)]}{W[x_1(s),x_2(s)]}f(s) \mathrm ds
$$

$$
W_1[x_1(s),x_2(s)]=
\begin{vmatrix}
0 & x_2(s)\\
1 & x'_2(s)
\end{vmatrix}=-x_2(s)\\
W_2[x_1(s),x_2(s)]=
\begin{vmatrix}
x_2(s) & 0\\
x'_1(s) & 1
\end{vmatrix}=x_1(s)\\
$$

因此, 当 $n=2$ 时, 常数变易公式变为

$$
\varphi(t)=\int_{t_0}^t\frac{x_2(t)x_1(s)-x_1(t)x_2(s)}{W[x_1(s),x_2(s)]}f(s)\mathrm ds
$$

而通解就是

$$
x=c_1x_1(t)+c_2x_2(t)+\varphi(t)
$$

这里 $c_1,c_2$ 是任意常数.

**例 3** 试求方程 
$$
x"+x=\tan t
$$

的一个解.

**解 ： ** 易知对应的齐次线性微分方程 $x"+x=0$ 的基本解组为 
$$
x_1(t)+x_2(t)=\cos t,x_2(t)=\sin t 
$$

我们利用公式来求解方程. 这时

$$
W[X_1(t),x_2(t)]=
\begin{vmatrix}
\cos t & \sin t \\
-\sin t & \cos t
\end{vmatrix}
=1
$$

$$
\begin{align}
\varphi (t) & =\int_0^t (\sin t \cos s - \cos t \sin s) \tan s \mathrm ds\\
& =\sin t \int_0^t\sin t \mathrm ds - \cos t \int_0^t\sin s \tan s \mathrm ds \\
& = \sin t(1-\cos t)+\cos t(\sin t + \ln |\sec t + \tan t|)\\
& = \sin t - \cos t\ln |\sec t+ \tan t|
\end{align}
$$ 

因为 $\sin t$ 是对应的齐次线性微分方程的一个解, 所以函数

$$
\bar{\varphi}=-\cos t \ln|\sec t + \tan t|
$$

也是原方程的一个解.



### 消元法: 求齐次线性方程的解的另一解法

保留一个未知函数 $x_1$, 消掉另一未知函数 $x_2$

$$
\begin{cases}
x'_1=x_1+x_2\\
x'_2=x_2
\end{cases}\\
x_1" = x'_1+x'_2 \quad x{_1}"=2x'_1-x_1\\
x_1"-2x'_1+x_1=0 \quad x_1=c_1e^t+c_2te^t\\
x_1"=x'_1-x_1\\
x_2=c_1e^t+c_2e^t+c_2te^t-c_1e^t-c_2te^t\\
x_2=c_2e^t
$$

于是

$$
\begin{cases}
x_1 =c_1e^t+c_2te^t\\
x_2 =c_2e^t
\end{cases}
$$

### 消元法: 求非齐次线性方程的解的另一解法

保留一个未知函数 $x_1$, 消掉另一未知函数 $x_2$

$$
\begin{cases}
x'_1=x_1+x_2+e^{-t}\\
x'_2=x_2
\end{cases}\\
x_1" = x'_1+x'_2-e^{-t} \\
x{_1}"=x'_1-x_2-e^{-t}\\
x_2=x'_1-x_1-e^{-t}\\
x_1"=2x'_1-x_1-2e^{-t} \\
x_1"-2x'_1+x_1=-2e^{-t}\\
x_1=c_1e^t+c_2te^t-\frac{1}{2}e^{-t}\\
x_2=c_1e^t+c_2e^t+c_2te^t+\frac{1}{2}e^{-t}-c_1e^t-c_2te^t+\frac{1}{2}e^{-t}-e^{-t}\\
x_2=c_2e^t
$$

$$
\begin{cases}
x_1=c_1e^t+c_2te^t-\frac{1}{2}e^{-t}\\
x_2=c_2e^t
\end{cases}\quad
x(0)=\begin{bmatrix}
-1\\
1
\end{bmatrix}\\
$$

$$
\begin{cases}
x_1=te^t-\frac{1}{2}(e^t+e^{-t})\\
x_2=e^t
\end{cases}\quad
\begin{bmatrix}
c_1\\
c_2
\end{bmatrix}=
\begin{bmatrix}
-\frac{1}{2}\\
1
\end{bmatrix}
$$
